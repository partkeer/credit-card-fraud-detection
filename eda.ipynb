```python
# Perform exploratory data analysis on dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv('creditcard.csv')

# Separate data into positive and negative classes
pos = data[data['Class'] == 1]
neg = data[data['Class'] == 0]

# inspect data
data.head()
```

```python
# statistical summary
data.describe()
```

```python
# inspect duplicate rows
data[data.duplicated(keep=False)]
```

```python
# pairplot of subset of negative samples
sns.pairplot(neg.iloc[:100, :5])
plt.suptitle('Pairplot of Negative Samples', fontweight='heavy', color='red')
plt.show()

# heatmap of correlations for negative samples
neg_corr = neg.corr()
sns.heatmap(neg_corr, cmap='coolwarm')
plt.show()
```

```python
# pairplot of subset of positive samples
sns.pairplot(pos.iloc[:100, :5])
plt.suptitle('Pairplot of Positive Samples', fontweight='heavy', color='red')
plt.show()

# heatmap of correlations for positive samples
pos_corr = pos.corr()
sns.heatmap(pos_corr, cmap='coolwarm')
plt.show()
```
